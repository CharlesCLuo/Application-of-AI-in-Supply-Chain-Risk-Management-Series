```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np



url = "https://raw.githubusercontent.com/CharlesCLuo/Application-of-AI-in-Supply-Chain-Risk-Management-Series/main/Demand_Forecsting/comprehensive_supply_chain_data_with_weather_spike.csv"
df = pd.read_csv(url, parse_dates=['Timestamp'], index_col='Timestamp')







plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Demand'], color='blue')
plt.title('Figure 8.1: Demand Over Time')
plt.xlabel('Date')
plt.ylabel('Demand')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()





plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Temperature (째C)'], color='orange')
plt.title('Figure 8.2: Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (째C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()





plt.figure(figsize=(7, 5))
sns.scatterplot(x=df['Temperature (째C)'], y=df['Demand'], hue=df['Weather Spike'], palette='coolwarm')
plt.title('Figure 8.3: Temperature vs. Demand')
plt.xlabel('Temperature (째C)')
plt.ylabel('Demand')
plt.grid(True)
plt.legend(title='Weather Spike', labels=['No Spike', 'Spike'])
plt.show()





plt.figure(figsize=(7, 5))
sns.barplot(x='Weather Spike', y='Demand', data=df, estimator=np.mean, palette='Set2')
plt.title('Figure 8.4: Average Demand During Weather Spikes vs. Normal Days')
plt.xlabel('Weather Spike (0: Normal, 1: Spike)')
plt.ylabel('Average Demand')
plt.grid(True)
plt.show()






plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Figure 8.5: Correlation Heatmap of Variables')
plt.show()
```
